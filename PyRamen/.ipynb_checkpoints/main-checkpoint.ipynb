```python
### -*- coding: UTF-8 -*-
"""PyRamen Homework Starter."""

# @TODO: Import libraries
import csv
from pathlib import Path

# @TODO: Set file paths for menu_data.csv and sales_data.csv
menu_filepath = Path('Resources/menu_data.csv')
sales_filepath = Path('Resources/sales_data.csv')

# @TODO: Initialize list objects to hold our menu and sales data
menu = []
sales = []

# @TODO: Read in the menu data into the menu list
with open(menu_filepath, 'r') as csvfile:
    # Pass in the csv file to the csv.reader() function
    # (with ',' as the delmiter/separator) and return the csvreader object
    csvreader = csv.reader(csvfile, delimiter=',')
    # Go to the next row from the start of the file (which is often the first row/header) and iterate line_num by 1
    header = next(csvreader)
    # Read each row of data after the header
    for row in csvreader:
        # Setting temporary variable equal to the value in the 2nd row and appending to menu variable
        menu_str = row
        # Append the row date value to the list of Dates
        menu.append(menu_str)
        
# @TODO: Read in the sales data into the sales list
with open(sales_filepath, 'r') as csvfile:
    # Pass in the csv file to the csv.reader() function
    # (with ',' as the delmiter/separator) and return the csvreader object
    csvreader = csv.reader(csvfile, delimiter=',')
    # Go to the next row from the start of the file (which is often the first row/header) and iterate line_num by 1
    header = next(csvreader)
    # Read each row of data after the header
    for row in csvreader:
        # Setting temporary variable equal to the value in the 2nd row and appending to sales variable
        sales_str = row
        # Append the row date value to the list of Dates
        sales.append(sales_str)

# @TODO: Initialize dict object to hold our key-value pairs of items and metrics
report = {}

# Initialize a row counter variable
rc=1
# Line_Item_ID,Date,Credit_Card_Number,Quantity,Menu_Item
Menu_Item=[]
Quantity=[]
Item=[]
Price=[]
Cost=[]
# @TODO: Loop over every row in the sales list object
for sale in sales:
    # @TODO: Initialize sales data variables
    Menu_Item.append(sale[4]) 
    Quantity.append(sale[3])
    # @TODO:
    # If the item value not in the report, add it as a new entry with initialized metrics
    # Naming convention allows the keys to be ordered in logical fashion, count, revenue, cost, profit
    if 'sale[4]'not in report.items():
        report[sale [4]]={"01-count":0,"02-revenue":0,"03-cogs":0,"04-profit":0}

  
    # @TODO: For every row in our sales data, loop over the menu records to determine a match

for n in menu:   
    for sale in sales:    
        Item.append(n[0])
        Price.append(n[3])
        Cost.append(n[4])
        if sale[4] == n[0]:
            # @TODO: Print out matching menu data <--- not doing. Output too messy.
            
            
            # @TODO: Cumulatively add up the metrics for each item key
            quantity=float(sale[3])
            price=float(n[3])
            cost=float(n[4])
            
            revenue=price*quantity
            cogs=cost*quantity
            profit=revenue-cogs
            
            report[sale[4]]["01-count"] += quantity
            report[sale[4]]["02-revenue"] += revenue
            report[sale[4]]["03-cogs"] += cogs
            report[sale[4]]["04-profit"] += profit         
        else:
            # @TODO: Else, the sales item does not equal any fo the item in the menu data, therefore no match
            # @TODO: Increment the row counter by 1
            # print(f"{rc}- {n[0]} does not equal {sale[4]}! NO MATCH!") <----- Commenting out. Output too messy.
            rc = rc+1
              
# @TODO: Print total number of records in sales data            
print(report)            
            
# @TODO: Write out report to a text file (won't appear on the command line output)  
ramen_type=[]
quantity=[]
revenue=[]
cogs=[]
profit=[]
rows=[]
# Set the output file path
output_path = Path('output.csv')
# Open the output path as a file object
with open(output_path, 'w') as csvfile:
    # Set the file object as a csvwriter object
    csvwriter = csv.writer(csvfile, delimiter=',')
    # Write the header to the output file
    title = ["Ramen Type", "Quantity", "Revenue $","COGS $","Profit $"]
    csvwriter.writerow(title)
    for row in report:
        # Write the list of metrics to the output file
        ramen_type.append(row)
        quantity.append(report[row]["01-count"])
        revenue.append(report[row]["02-revenue"])
        cogs.append(report[row]["03-cogs"])
        profit.append(report[row]["04-profit"])
    for j in range(0,len(ramen_type)):
        dummy=[ramen_type[j],quantity[j],revenue[j],cogs[j],profit[j]]        
        csvwriter.writerow(dummy)
# END
```
