```python
# Python script that analyses the financial records of your company to calculate each of the following:
# The total number of months included in the dataset
# The net total amount of Profit/Losses over the entire period
# The average of the changes in Profit/Losses over the entire period
#The greatest increase in profits (date and amount) over the entire period
#The greatest decrease in losses (date and amount) over the entire period

# Step 1: Importing the pathlib and csv library for data input
from pathlib import Path
import csv
# Setting the file path for data input
csvpath = Path('Resources/budget_data.csv')
# Initializing variables to hold dates and Profit/Losses
dates = []
profit_loss = []
# Opening the input path as a file object
with open(csvpath, 'r') as csvfile:

    # Pass in the csv file to the csv.reader() function
    # (with ',' as the delmiter/separator) and return the csvreader object
    csvreader = csv.reader(csvfile, delimiter=',')
    # Go to the next row from the start of the file (which is often the first row/header) and iterate line_num by 1
    header = next(csvreader)
    # Read each row of data after the header
    for row in csvreader:
        # Setting date variable equal to the value in the 1st column of each row and profit/loss variable to 2nd column
        date_str = str(row[0])
        pro_loss = float(row[1])
       # Append the row date value to the list of Dates
        dates.append(date_str)
        profit_loss.append(pro_loss)
# Calculating:
# The total number of months included in the dataset
no_months=len(dates)
# The net total amount of Profit/Losses over the entire period
net_total_pro_loss=sum(profit_loss)
# The average of the changes in Profit/Losses over the entire period
change=[]
for n in range(1,len(profit_loss)):
    delta=profit_loss[n]-profit_loss[n-1]
    change.append(delta)
ave_pro_loss=round(sum(change)/len(change),2)
#The greatest increase in profits (date and amount) over the entire period
max_inc=0
i=0
for m in range(0,len(change)):
    if change[m]>=max_inc:
        max_inc=change[m]
        i=m
#The greatest decrease in losses (date and amount) over the entire period
max_dec=0
for o in range(0,len(change)):
    if change[o]<=max_dec:
        max_dec=change[o]
        j=o
# Printing Output 
# In Terminal
print("_____________________________________________________________")
print(f"Total Months: {no_months}\nTotal: ${net_total_pro_loss}\nAverage  Change: ${ave_pro_loss}\nGreatest Increase in Profits: {dates[i]} (${max_inc})\nGreatest Decrease in Profits: {dates[j]} (${max_dec})")
# In File
# Creating a list of results
max_inc=[max_inc,dates[i]]
Max_Dec=[max_dec,dates[j]]
title = ["Total Months", "Total $", "Average  Change $","Greatest Increase in Profits $","Greatest Decrease in Profits $"]
results = [no_months, net_total_pro_loss, ave_pro_loss,max_inc,max_dec]

# Set the output file path
output_path = Path('output.csv')

# Open the output path as a file object
with open(output_path, 'w') as csvfile:
    # Set the file object as a csvwriter object
    csvwriter = csv.writer(csvfile, delimiter=',')
    # Write the header to the output file
    csvwriter.writerow(title)
    # Write the list of metrics to the output file
    csvwriter.writerow(results)

  
```
